###### Creating a Age column for the players - Not Required

```python
df['dob'].dtypes
```

```python
#Change DOB object into DateTime
df["dob"]= pd.to_datetime(df["dob"]) 
```

```python
df["dob"]
```

```python
df['dob'].dtypes
```

```python
df['Today'] = pd.to_datetime('today')
df['Today']
```

```python
df['Age'] = df['Today'] - df['dob']
```

```python
df['Age']
```

```python
#Converting the Days to Years
df['Age'] = df['Age']/np.timedelta64(1,'Y')
```

```python
df['Age']
```

# EDA

###### Distribution of Age - Histogram

```python
df.head(2)
```

```python
data = [go.Histogram(
        x = df['Age'],
        name='Age',
        xbins=dict(size=3)
)]

layout = go.Layout(
    title = "Age Distribution",
    xaxis = dict(title = 'Player Age Distribution')
)

fig = go.Figure(data=data, layout=layout)
fig.iplot()
```

###### Distribution of Wage and Age

```python
df.head(2)
```

```python
data = [go.Histogram(
        x = df[df['team_position']=='GK']['wage_eur'],
        name = 'Goalkeeper Wage',
        xbins = dict(end=10000)
),       
        go.Histogram(
        x = df[df['team_position']=='ST']['wage_eur'],
        name='Striker Wage',
        xbins = dict(end=10000)
)]

layout = go.Layout(
    title = "Overall Skill",
    xaxis = dict(title = 'Player Skill Distribution')
)

fig = go.Figure(data=data, layout=layout)
fig.iplot()
```

###### Overall Skill vs Player Value Bubble Plot

```python
df.head(2)
```

```python
data = [go.Scatter(
        x = df['value_eur'],
        y = df[df['overall'] > 75]['overall'],
        text=df['short_name'],
        mode='markers',
        
        marker=dict(size=df['age']/2,
        color = df['age'],
        showscale = True))]

layout = go.Layout(
    title = "Skill vs Value Comparison",
    xaxis = dict(title = 'Player Value'),
    yaxis = dict(title = 'Player Skill')
    
)

fig = go.Figure(data=data, layout=layout)
fig.iplot()
```

###### Bar Plot Comparing Players under the age of 23 whose current skill is less than 70 and their potential to reach over 85

```python
df.head(2)
```

```python
youngpotentialtop10 = df[(df['age'] < 22) & (df['potential'] > 85) & (df['overall'] < 70)].head(20)
youngpotentialtop10.head()
```

```python
overall = go.Bar(
    x = youngpotentialtop10['short_name'],
    y = youngpotentialtop10['overall'],
    name = 'Overall Skill'
)

potential = go.Bar(
    x = youngpotentialtop10['short_name'],
    y = youngpotentialtop10['potential'],
    name = 'Potential Skill'
)

data = [overall, potential]

layout = go.Layout(
    title = 'Player Potential for U-23 Age Group'
)

fig = go.Figure(data=data, layout=layout)
fig.iplot()
```

###### Average Wages of popular 5 clubs

```python
df.head(1)
```

```python
barcelona = df[df['club'] == 'FC Barcelona']['wage_eur']
realmadrid = df[df['club'] == 'Real Madrid']['wage_eur']
manchesterunited = df[df['club'] == 'Manchester United']['wage_eur']
liverpool = df[df['club'] == 'Liverpool']['wage_eur']
paris = df[df['club'] == 'Paris Saint-Germain']['wage_eur']
```

```python
data = [
    go.Box(
        y=barcelona,
        name='Barcelona',
        boxpoints='all', # display the original data points
        jitter=0.9,      # spread them out so they all appear
        pointpos=-1.8,
        text=df[df['club'] == 'FC Barcelona']['short_name']
    ),
    go.Box(
        y=realmadrid,
        name='Real Madrid',
        boxpoints='all', # display the original data points
        jitter=0.9,      # spread them out so they all appear
        pointpos=-1.8,
        text=df[df['club'] == 'Real Madrid']['short_name']
    ),
    go.Box(
        y=manchesterunited,
        name='Manchester United',
        boxpoints='all', # display the original data points
        jitter=0.9,      # spread them out so they all appear
        pointpos=-1.8,
        text=df[df['club'] == 'Manchester United']['short_name']
    ),
    go.Box(
        y=liverpool,
        name='Liverpool',
        boxpoints='all', # display the original data points
        jitter=0.9,      # spread them out so they all appear
        pointpos=-1.8,
        text=df[df['club'] == 'Liverpool']['short_name']
    ),
    go.Box(
        y=paris,
        name='Paris Saint-Germain',
        boxpoints='all', # display the original data points
        jitter=0.9,      # spread them out so they all appear
        pointpos=-1.8,
        text=df[df['club'] == 'Paris Saint-Germain']['short_name']
    )
]

layout = go.Layout(
    title = 'Wage Comparison between Clubs'
)

fig = go.Figure(data=data, layout=layout)
fig.iplot()
```
